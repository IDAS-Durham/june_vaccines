```python
import numpy as np
import pandas as pd
from pathlib import Path
from june.records import RecordReader
import datetime
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

plt.style.use(['science'])
plt.style.reload_library()
```

```python
target_az = 0.67
target_az_minus = 0.62
target_az_plus = 0.71
target_pfizer = 0.8
target_pfizer_minus = 0.77
target_pfizer_plus = 0.83
# https://www.medrxiv.org/content/10.1101/2021.08.18.21262237v1.full.pdf
```

```python
susceptibilities = 1 - np.linspace(0.8,1.,20)
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(9,3),sharey=True)
for run in range(20):
    az = pd.read_csv()# Dataset not publicly available, please contact the authors to be granted access
    az['date'] = pd.to_datetime(az['date'])
    az.set_index('date',inplace=True)
    az = az.iloc[1:]
    az.columns = ['']
    if run == 19:
        az = az.replace(np.nan, 1.)
    az.plot(
            ax=ax[0],
           legend=False)

for run in range(20):
    pfizer = pd.read_csv()# Dataset not publicly available, please contact the authors to be granted access
    pfizer['date'] = pd.to_datetime(pfizer['date'])
    pfizer.set_index('date',inplace=True)    
    pfizer.columns = [f'{susceptibilities[run]:.2f}']
    pfizer = pfizer.iloc[1:]
    if run == 19:
        pfizer = pfizer.replace(np.nan, 1.)
    pfizer.plot(label=f'{susceptibilities[run]:.2f}',
                legend=False,
                ax=ax[1])

            
    
ax[0].fill_between(az.index, target_az_minus, target_az_plus, label='Observed VE', alpha=0.2)
ax[1].fill_between(pfizer.index, target_pfizer_minus, target_pfizer_plus, label='Observed VE CI', alpha=0.2)
ax[0].axhline(y=target_az, linestyle='dashed')
ax[1].axhline(y=target_pfizer, linestyle='dashed',label='Observed VE')
ax[1].legend(bbox_to_anchor=(1,1),title='Susceptibility',ncol=2)
ax[0].set_xlabel('')
ax[1].set_xlabel('')
ax[0].set_title('AstraZeneca')
ax[1].set_title('Pfizer')
ax[0].set_ylabel('$VE_\mathrm{infection}$')
ax[1].set_ylabel('$VE_\mathrm{infection}$')

#ax.set_ylabel('AZ Vaccine Efficacy')
plt.savefig('figures/vaccine_efficacy.pdf')
```
